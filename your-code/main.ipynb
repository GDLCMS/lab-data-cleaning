#### 3. Create a mysql engine to set the connection to the server. Check the connection details in [this link](https://relational.fit.cvut.cz/dataset/Stats)

```python
# hostname: relational.fit.cvut.cz
#port: 3306
#username: guest
#password: relational
#database: stats

engine = create_engine('mysql+pymysql://relational.fit.cvut.cz:3306/stats?user=guest&password=relational')

try:
    connection = engine.connect()
    print("Connection established")
except:
    print("Connection failed")
    
```

#### 4. Import the users table 

```python
users = pd.read_sql_query('SELECT * FROM users', engine)
```

#### 5. Rename Id column to userId

```python
users.rename(columns={'Id':'userId'}, inplace=True)
```

#### 6. Import the posts table. 

```python
posts = pd.read_sql_query('SELECT * FROM posts', engine)
```

#### 7. Rename Id column to postId and OwnerUserId to userId

```python
posts.rename(columns={'Id':'postId', 'OwnerUserId':'userId'}, inplace=True)
posts.head()
```

#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userID,ViewCount,CommentCount

```python
users = users[['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']]
posts = posts[['postId', 'Score', 'userId', 'ViewCount', 'CommentCount']]
users.head()
```

```python
posts.head()
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
posts_with_users = pd.merge(users, posts, on='userId')
posts_with_users.head()
```

#### 9. How many missing values do you have in your merged dataframe? On which columns?

```python
posts_with_users.isnull().sum()
```

#### 10. You will need to make something with missing values.  Will you clean or filling them? Explain.

```python
posts_with_users.fillna(0, inplace=True)
posts_with_users.head()
```

```python
posts_with_users.dtypes
posts_with_users['ViewCount'] = posts_with_users['ViewCount'].astype(int)
posts_with_users['CommentCount'] = posts_with_users['CommentCount'].astype(int)
posts_with_users['userId'] = posts_with_users['userId'].astype(int)
posts_with_users['postId'] = posts_with_users['postId'].astype(int)
posts_with_users['Reputation'] = posts_with_users['Reputation'].astype(int)
posts_with_users['Views'] = posts_with_users['Views'].astype(int)
posts_with_users['UpVotes'] = posts_with_users['UpVotes'].astype(int)
posts_with_users['DownVotes'] = posts_with_users['DownVotes'].astype(int)
posts_with_users['Score'] = posts_with_users['Score'].astype(int)
```

#### Bonus: Identify extreme values in your merged dataframe as you have learned in class, create a dataframe called outliers with the same columns as our data set and calculate the bounds. The values of the outliers dataframe will be the values of the merged_df that fall outside that bounds. You will need to save your outliers dataframe to a csv file on your-code folder.

```python
posts_with_users.describe()
```

```python
posts_with_users.describe().T
```

```python
outliers = posts_with_users[['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes', 'postId', 'Score', 'ViewCount', 'CommentCount']]
outliers.describe()
```

```python
Q1 = outliers.quantile(0.25)
Q3 = outliers.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
posts_with_users.shape
```

```python
posts_with_users_outliers = outliers[~((outliers < (Q1 - 1.5 * IQR)) |(outliers > (Q3 + 1.5 * IQR))).any(axis=1)]
posts_with_users_outliers.shape
```

```python
posts_with_users_outliers.to_csv('outliers2.csv', index=False)
```
